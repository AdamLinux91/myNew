```python
import cvzone
from ultralytics import YOLO
import cv2
import math
#import pygame
import serial 

cap = cv2.VideoCapture(r"WhatsApp Video 2025-04-20 at 10.16.23_a744e891.mp4")  

model = YOLO(r"C:\Users\Dream\Desktop\asmaa\Fire\Fire\Fire\Fire\fire.pt")

ser=serial.Serial('COM6',9600)

while True:
    ret, frame = cap.read()
    frame = cv2.resize(frame, (640, 480))
    result = model(frame, stream=True)

    for info in result:
        boxes = info.boxes
        for box in boxes:
            confidence = box.conf[0]
            confidence = math.ceil(confidence * 100)
            
            if confidence > 40:
                x1, y1, x2, y2 = box.xyxy[0]
                x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 0, 255), 5)
                cvzone.putTextRect(frame, f'Fire {confidence}%', [x1 + 8, y1 + 100],
                                   scale=1.5, thickness=2)
                ser.write(b'1')

              
            else:
                ser.write(b'0')
                pass  

    cv2.imshow('fier', frame)
    if cv2.waitKey(1) == ord('q'):
        break


cap.release()
cv2.destroyAllWindows()
```
